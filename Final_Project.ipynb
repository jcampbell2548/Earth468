```python
day_of_year = np.array(df_1960s['doy'])
daily_mean = np.array(df_1960s['day_mean'])
cost_unc = np.std(np.array(df_1960s['day_mean']))
cost_unc
```

```python
def costs(day, A, lamda): # AC is data context, A and lamda are model state parameters
    return A/np.cosh(day/lamda) # np.cosh is hyperbolic cosecant
```

```python
def L1(data,pred,unc):
    return np.sum(np.abs(data-pred)/unc,axis=1)
```

```python
A_range = [-50,50]
lamda_range = [0,366]

N = 1*10**4

As = np.random.uniform(A_range[0],A_range[1],N)
lamdas = np.random.uniform(lamda_range[0],lamda_range[1],N)

trials = np.array([As,lamdas]).T
trials
```

```python
predictions = []
for i in range(N):
    preds = costs(day_of_year, trials[i,0], trials[i,1])
    predictions.append(preds)
    
predictions = np.array(predictions) # convert to array
np.shape(predictions)

L1_misfits = L1(daily_mean, predictions, cost_unc)
L1_misfits
```

```python
index_bestsolution = np.where(L1_misfits==min(L1_misfits))[0]
# L1_misfits[index_bestsolution],min(L1_misfits) Checks that index corresponds to lowest misfit, ie best solution
best_model = trials[index_bestsolution][0]
best_predictions = predictions[index_bestsolution]
best_predictions
```

```python
fig = plt.figure(figsize=[8,8])
ax = fig.add_subplot(111)
ax.errorbar(day_of_year,daily_mean,yerr=cost_unc,linestyle='None',marker='o',label='observations',color='cyan')

ax.plot(day_of_year,best_predictions.T,color='magenta',label='model prediction')

ax.set_xlabel('% of households with central AC',size=15)
ax.set_ylabel('cost as % of monthly income',size=15)

ax.set_title('best model fit for A={} and λ={}'.format(round(best_model[0],2),round(best_model[1],2)),size=20)

ax.legend();
```

```python
def G_line(context):
    G = np.ones((len(context),2))
    G[:,1] = context
    return G

# Since we are still dealing with data that has missing values, we will want to create masks for 
# each station that will only consider data context and data when a measurement exists. 
# This is an example for a single column/station of data:





data_context = doy_1960
data = means_1960

G = G_line(data_context) #create G matrix for station_index

LS = np.linalg.inv(G.T@G)@G.T@data #Least squares solution for station_index
print('best fit intercept:',LS[0],'\nbest fit slope:',LS[1])

predictions = G@LS #compute predictions, should be as many values as snowfall[mask,station_index]

residuals = predictions - data

plt.scatter(data_context,residuals)
LS
```

```python
array_1960s = np.array(df_1960s)

from scipy.stats import pearsonr
corrcoeff_1960s, z = pearsonr(array_1960s[:,1],array_1960s[:,2])

def bootstrap_resampling(obs):
    
    resampled = [] # empty list
    n = obs.shape[0] # number of rows for the input array
    
    for i in np.arange(n): # iterates through n times
        random_index = np.random.randint(0,obs.shape[0])
        random_row = obs[random_index]
        resampled.append(random_row)
        
    return np.array(resampled)

bootstrap_resampling(array_1960s)
   

```

```python
cc_1960s = [] # empty lists need to be outside the for loop so the values save in the list each time
    
for i in np.arange(10000):
    resampled = bootstrap_resampling(array_1960s)
    corrcoeff_1960s, z = pearsonr(resampled[:,1],resampled[:,2]) 
    cc_1960s.append(corrcoeff_1960s)
    
cc_1960s_array = np.array(cc_1960s)
```

```python
fig = plt.figure(figsize=[4,4])
ax = fig.add_subplot(111)
ax.hist(cc_1960s_array,bins=50,color='steelblue')
ax.set_xlabel('Correlation Coefficients')
ax.set_ylabel('Counts')
ax.set_title('1960s')
plt.show()
```
