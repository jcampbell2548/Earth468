```python
import numpy as np
import matplotlib.pyplot as plt
A = np.random.uniform(3.2, 8.8, 25)
```

```python
fig = plt.figure(figsize=[4,4])
ax = fig.add_subplot(111)
ax.hist(A,bins=5,color='red')
ax.set_xlabel('Randomly Generated Number')
ax.set_ylabel('Counts')
ax.set_title('Vector A Histogram')
plt.show() 
```

```python
D = np.random.normal(80, 2, 1000)
δ = np.random.normal(1.5, 0.1, 1000)
```

```python
fig = plt.figure(figsize=[4,4])
ax = fig.add_subplot(111)
ax.hist(D,bins=50,color='green')
ax.set_xlabel('Randomly Generated Number')
ax.set_ylabel('Counts')
ax.set_title('D Histogram')
plt.show() 
```

```python
fig = plt.figure(figsize=[4,4])
ax = fig.add_subplot(111)
ax.hist(δ,bins=50,color='purple')
ax.set_xlabel('Randomly Generated Number')
ax.set_ylabel('Counts')
ax.set_title('δ Histogram')
plt.show() 
```

```python
heights = []
def heights(D,δ):
    d = 2*D*np.tan((δ*np.pi/180)/2)
    return d
```

```python
yeti_heights = heights(D,δ)
print(yeti_heights.shape)
```

```python
fig = plt.figure(figsize=[4,4])
ax = fig.add_subplot(111)
ax.hist(yeti_heights,bins=50,color='pink')
ax.set_xlabel('Randomly Generated Number')
ax.set_ylabel('Counts')
ax.set_title('d Histogram')
plt.show() 
```

```python
np.mean(yeti_heights)
```

```python
np.std(yeti_heights)
```

```python
np.median(yeti_heights)
```

```python
np.percentile(yeti_heights,[5,95])
```

The mean of the heights is about 2.09 meters, the standard deviation is about 0.15 meters, the median is about 2.09 meters, and the 90% confidence interval is (1.86,2.34).
