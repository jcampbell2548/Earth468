## <font color='red'>Import Libraries: </font>

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## <font color='red'>a) 2 pts</font>

Load the vectors and plot each of them with circle markers to aid in recognition of what shape the vectors outline.

```python
df = pd.read_csv('Vectors.csv',header=None)
vectors = np.array(df).T
```

```python
fig = plt.figure(figsize=[5,5])
ax = fig.add_subplot(111)

ax.plot(vectors[0],vectors[1],marker='o')

ax.set_xlim(-2,2)
ax.set_ylim(-2,2)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Original Pikachu')

plt.show()
```

## <font color='red'>b) 4 pts</font>

Map each of the 120 vectors using the matrix

$$A = \begin{pmatrix} 1.00 & 1.50 \\ 0.00 & 1.00 \end{pmatrix} $$

Be sure to right multiply (i.e., Av).

Plot the new vectors and describe the perspective of the new space. <b> In other words, how is the outline of
the image formed from connecting the vectors different from the original? </b> It might be helpful to include your original plot in the figure to compare the new vectors to the original vectors.

```python
A = np.array([[1,1.5],[0,1]])
A.shape
A
partb = np.dot(A,vectors)
```

```python
fig = plt.figure(figsize=[5,5])
ax = fig.add_subplot(111)

ax.plot(vectors[0],vectors[1],marker='o')
ax.plot(partb[0],partb[1],marker='o')

ax.set_xlim(-2,2)
ax.set_ylim(-2,2)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Vector A Pikachu')

plt.show()
```

The outline of the image formed from connecting these vectors is different from the original outline because the Pikachu is deformed, it looks stretched and diagonal. So, it takes up more space on the x axis but about the same amount of space on the y axis as the original.

## <font color='red'>d) 6 pts</font>

Compute the area of the polygon defined by connecting the original
120 vectors, as well as the area of the shape mapped under matrix A and matrix B.

<b> What is the ratio of the area of the shape mapped under A and B to the original area? </b>

Hint: To compute the area of a polygon defined by a collection of (x,y) vertices, find a function on stack overflow that performs this calculation for us
(https://stackoverflow.com/questions/24467972/calculate-area-of-polygon-given-x-y-coordinates)


```python
# Original Area
x = vectors[0]
y = vectors[1]
def PolyArea(x,y):
    return 0.5*np.abs(np.dot(x,np.roll(y,1))-np.dot(y,np.roll(x,1)))
print(PolyArea(x,y))
```

```python
# A Area
x = partb[0]
y = partb[1]
def PolyArea(x,y):
    return 0.5*np.abs(np.dot(x,np.roll(y,1))-np.dot(y,np.roll(x,1)))
print(PolyArea(x,y))
```

```python
# B Area
x = partc[0]
y = partc[1]
def PolyArea(x,y):
    return 0.5*np.abs(np.dot(x,np.roll(y,1))-np.dot(y,np.roll(x,1)))
print(PolyArea(x,y))
```

```python
print("The ratio of the area of the shape mapped under A to the original area is", 2.140247193989623/2.1402471939896253) 
```

```python
print("The ratio of the area of the shape mapped under B to the original area is", 4.6229339390175905/2.1402471939896253) 
```

## <font color='red'>e) 8 pts</font>

The determinant is a fundamental quantity of any square matrix (a square matrix is one with the same
number of rows as columns; square matrices map from an n-dimension space to another n-dimension
space). In python the determinant of matrix A is computed using numpy's linear algebra library, *np.linalg*, as

np.linalg.det(A)

<b> Compute the determinant of the matrix A and B. Based on the results, what is the significance of the
determinant of a matrix? (Hint: You might want to compare to the answers in part (d).) </b>

```python
np.linalg.det(A) # Determinant of matrix A
```

```python
np.linalg.det(B) # Determinant of matrix B
```

Based on these results, which are the same as the results from part (d), the significance of the determinant of a matrix is the extent of the transformation or deformation on the vectors. Since these matrices are 2x2, the determinant tells us about the area change. If it were a 3x3 matrix, the determinant would tell us about the volume change.

## <font color='red'>g) 6 pts</font>

Map each of the original vectors by AB and BA. Plot both sets of transformed vectors. 

<b> Describe the resulting transformations to the original shape. Is the result of the two mappings the same? If different, discuss why. </b>

```python
AB_array = np.array(AB)
BA_array = np.array(BA)
AB_partg = np.dot(AB_array,vectors)
BA_partg = np.dot(BA_array,vectors)

fig = plt.figure(figsize=[5,5])
ax = fig.add_subplot(111)

ax.plot(vectors[0],vectors[1],marker='o')
ax.plot(AB_partg[0],AB_partg[1],marker='o')
ax.plot(BA_partg[0],BA_partg[1],marker='o')

ax.set_xlim(-4,4)
ax.set_ylim(-2,2)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('AB and BA Pikachu')

plt.show()
```

The resulting transformations are stretched and deformed from the original shape. The result of the two mappings are not the same, since the vectors are different (because the order of multiplication matters so AB is not the same as BA).

## <font color='red'>h) 6 pts</font>
Construct a matrix which flips the shape along the vertical. Plot both the original and the flipped vectors.

<b> What is the determinant of this matrix? </b>

```python
flip_array = np.array([[-1,0],[0,1]])
flippy = np.dot(flip_array,vectors)

fig = plt.figure(figsize=[5,5])
ax = fig.add_subplot(111)

ax.plot(vectors[0],vectors[1],marker='o')
ax.plot(flippy[0],flippy[1],marker='o')

ax.set_xlim(-2,2)
ax.set_ylim(-2,2)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Flipped Pikachu')

plt.show()
```

```python
print("The determinant of this matrix is", np.linalg.det(flip_array))
```
